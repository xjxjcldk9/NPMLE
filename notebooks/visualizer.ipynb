```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
priors = ['uniform', 'normal', 'beta']
second_stages = ['linear', 'square', 'log', 'cross']
```

```python
def show_result(pior, second_stage):
    fig, axs = plt.subplots(2,2, figsize=(18,6))
    source = Path('dim2_combinations') / f'{prior}_{second_stage}'

    axs[0,0].axis('off')
    df1 = pd.read_csv(source/'beta_pm.csv', index_col='n')
    df1 = np.round(df1,3)

    pd.plotting.table(axs[0,0], df1, loc='center', colWidths=[0.2]*3)

    axs[0,0].set_title('beta posteriror mean')

    axs[0,1].axis('off')
    df2 = pd.read_csv(source/'prior.csv', index_col=0)
    df2 = np.round(df2,3)

    pd.plotting.table(axs[0,1], df2, loc='center', colWidths=[0.2]*5)

    axs[0,1].set_title('prior estimation')



    with source / 'consistent.png' as image_file:
        image = plt.imread(image_file)
        
    axs[1,0].axis('off')
    axs[1,0].imshow(image)



    with source / 'prior_compare.png' as image_file:
        image = plt.imread(image_file)
        
    axs[1,1].axis('off')
    axs[1,1].imshow(image)

    fig.suptitle( f'{prior}_{second_stage}')
```

```python
for prior in priors:
    for second_stage in second_stages:
        show_result(prior, second_stage)
```
